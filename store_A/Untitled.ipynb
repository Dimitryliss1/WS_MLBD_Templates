# 1.2 Предобработка данных

Как видно, из-за разделения по запятой и пробелу в некоторых столбцах произошло смещение на один или ряд вправо. Для последующей обработки посмотрим, сколько таких вхождений. Для начала нужно понять, какие значения могут находиться в поле `unit`. Построим множество из всех значений этого поля, которые по длине не превосходят шести символов.

```python
units = set()

for i in file['unit']:
    if i.strip() and len(i.strip()) < 7 and (i.isalpha() or i.endswith('.') or (i[-1].isnumeric() and i[0].isalpha())):
        units.add(i.strip())
units
```

Как видно, нам подходят следующие значения:
- кг (с точкой и без)
- км
- компл (с точкой и без)
- л
- м (с точкой и без)
- п.м.
- рул (с точкой и без)
- т (с точкой и без)
- упак
- шт (с точкой и без)
- м2
- м3
- дм2

Все значения, которые не входят в эти, будем соединять с предыдущим столбцом.

```python
suitable = {'кг', 'кг.',
            'км',
            'компл.', 'компл', 'комлп.',
            'л',
            "м", "м.",
            "п.м.",
            "рул.", "рул",
            "т.", "т",
            "упак",
            "шт", "шт.",
            "м2", "м3", "дм2"
           }
```

```python
# file.iloc[1]['name'] = file.iloc[1]['name'] + ', ' + file.iloc[1]['unit']
a = file.iloc[1]
copy_a = list(a)
copy_a[2] = ', '.join(copy_a[2:4])
copy_a.pop(3)
copy_a.append(None)
a = pd.DataFrame(data=copy_a, index=a.keys(), columns=[1])
file.iloc[1] = a.transpose()
```

Выше приведен образец сокращения для одной ячейки. Нам нужно провести такую операцию для всех ячеек.

```python
biased = [1]
while biased:
    biased = []
    for i in range(len(file)):
        if file.iloc[i]['unit'] and file.iloc[i]['unit'].strip() not in suitable:
            a = file.iloc[i]
            copy_a = list(a)
            copy_a[2] = ', '.join(copy_a[2:4])
            copy_a.pop(3)
            copy_a.append(None)
            a = pd.DataFrame(data=copy_a, index=a.keys(), columns=[i])
            file.iloc[i] = a.transpose()
    for i in file['unit']:
        if i.strip() and i.strip() not in suitable:
            biased.append(i.strip())
```

Посчитаем в каждом столбце количество значений `NaN`. Если их больше определенного числа, удалим их

```python
perc_dict = {'column': [], 'percentage': []}
for i in file.columns:
    perc_dict['column'].append(i)
    perc_dict['percentage'].append((file[i].isna().sum() + (file[i].value_counts()[''] if '' in file[i].value_counts() else 0))/(len(file)) * 100)

show_perc = pd.DataFrame(perc_dict)
show_perc
```

Полностью из отсутствующих значений состоят поля `Unnamed: 26-34`. Удалим эти поля

```python
to_drop = ['salvo_per_unit_of_everything',
           'justification_',
           '_justification_',
           'total_', 
           'k_t', 
           'class', 
           'unit_kilogram',
           'general_',
           'titul_number'
          ]

for i in range(20, 35):
    to_drop.append(f"Unnamed: {i}")

file = file.drop(columns=to_drop)
file
```

Техническими данными в датасете являются поля `number_rsl`, `number_PP`, `number_slot`, `enum_cd_number`. Удалим эти данные

```python
to_drop_tech = ['number_rsl', 'number_PP', 'number_slot', 'enub_cd_number']
file = file.drop(columns=to_drop_tech)
```

```python
file
```

Поле `price` иногда содержит странные значения, вроде `2528,22\n19035/ндс/а*тзр`. В этих строках зачастую значащим является число, стоящее до первого `\n`.

```python
for i in range(len(file)):
    if type(file.iloc[i]['price']) != float:
        file.iloc[i]['price'] = file.iloc[i]['price'].split('\n')[0]
file
```

Преобразуем значения в столбцах `total_amount`, `price`, `total`, `salvo_per_unit_of_everything_` в тип `float` там, где это возможно.

```python
columns_to_transform = ['total_amount', 'price', 'total', 'salvo_per_unit_of_everything_']
for i in columns_to_transform:
    for j in range(len(file)):
        if type(file.iloc[j][i]) != float:
            try:
                file.iloc[j][i] = float(file.iloc[j][i].replace(",", "."))
            except ValueError:
                file.iloc[j][i] = None
```

Закодируем данные в столбце `unit` при помощи `LabelEncoder` из модуля `sklearn.preprocessing`, при этом сохраним классы для дальнейшего использования

```python
le = LabelEncoder()
file['unit'] = le.fit_transform(file['unit'])
print(dict(zip(range(len(le.classes_)),le.classes_)))
```

```python
import json

with open("classes.json", "w") as dump_file:
    json.dump(dict(zip(range(len(le.classes_)), le.classes_)), dump_file)
```
